```python
import ipywidgets as widgets
from IPython.display import display
import math

# Function to perform calculations
def calculate(expression):
    try:
        result = eval(expression)
        output_text.value = str(result)
    except Exception as e:
        output_text.value = 'Error'

# Function to handle button clicks
def on_button_click(b):
    if b.description == '=':
        calculate(input_text.value)
    elif b.description == 'C':
        input_text.value = ''
        output_text.value = ''
    else:
        input_text.value += b.description

# Function to simulate hover effect (change button color on hover)
def on_hover(btn):
    btn.style.button_color = '#FF1493'  # Darker hot pink on hover

# Function to revert color back after hover
def on_hover_out(btn):
    btn.style.button_color = btn.default_color  # Reset to original color

# Creating input fields for expression and result
input_text = widgets.Text(placeholder='Enter expression', layout=widgets.Layout(width='400px', height='40px'))
output_text = widgets.Text(value='', placeholder='Result', layout=widgets.Layout(width='400px', height='40px'))

# Create buttons for the calculator
button_layout = widgets.Layout(width='70px', height='70px',
                               border_radius='15px',  # Curved buttons
                               box_shadow='2px 2px 5px rgba(0, 0, 0, 0.2)',  # Button shadow
                               margin='5px')  # Space around buttons

# Style buttons with colors
number_button_style = {'button_color': '#0ABAB5'}  # Tiffany Blue
operator_button_style = {'button_color': '#FF69B4'}  # Hot Pink

# Define button labels and their styles
buttons = [
    ('7', number_button_style), ('8', number_button_style), ('9', number_button_style), ('/', operator_button_style), ('sqrt', operator_button_style),
    ('4', number_button_style), ('5', number_button_style), ('6', number_button_style), ('*', operator_button_style), ('pow', operator_button_style),
    ('1', number_button_style), ('2', number_button_style), ('3', number_button_style), ('-', operator_button_style), ('log', operator_button_style),
    ('0', number_button_style), ('.', number_button_style), ('=', operator_button_style), ('+', operator_button_style), ('C', operator_button_style)
]

# Create button widgets and set default colors for hover simulation
button_widgets = []
for label, style in buttons:
    btn = widgets.Button(description=label, layout=button_layout, style=style)
    btn.default_color = style['button_color']  # Save default color
    btn.on_click(on_button_click)  # Set button click handler
    btn.on_hover = lambda btn=btn: on_hover(btn)  # Bind hover event to change color
    btn.on_mouseout = lambda btn=btn: on_hover_out(btn)  # Reset color after hover out
    button_widgets.append(btn)

# Arrange buttons in a grid
grid = widgets.GridBox(button_widgets, layout=widgets.Layout(
    grid_template_columns='repeat(5, 70px)', grid_gap='10px'  # Space between buttons
))

# Display the input fields and buttons
display(input_text, output_text, grid)
```
